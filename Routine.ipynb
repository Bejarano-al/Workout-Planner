```python
#list of exercises  
skull_crushers =       ["-[5Kg]---------[Dumbbells]----[10 Reps - 5 Sets - 1m Rest ]","Skull Crushers"]
dips =                 ["-[Body Weight]-[Bars]---------[10 Reps - 5 Sets - 1m Rest ]","Dips"]
dumbbell_kickbacks =   ["-[5Kg]---------[Dumbbells]----[10 Reps - 5 Sets - 1m Rest ]","Dumbbell Kickbacks"]
diamond_pushups =      ["-[Body Weight]-[No equipment]-[5  Reps - 5 Sets - 1m Rest ]","Diamond Push-ups"]
bench_press =          ["-[16,5Kg]------[Dumbbells]----[10 Reps - 5 Sets - 1m Rest ]","Bench Press"]
bicep_curls =          ["-[11,5Kg]------[Dumbbells]----[10 Reps - 5 Sets - 1m Rest ]","Bicep Curls"]
hammer_curls =         ["-[11,2Kg]------[Dumbbells]----[10 Reps - 5 Sets - 1m Rest ]","Hammer Curls"]
chinups =              ["-[Body Weight]-[Bars]---------[5  Reps - 5 Sets - 30s Rest]","Chin-ups"]
neutral_grip_pullups = ["-[Body Weight]-[Bars]---------[5 Reps  - 5 Sets - 30s Rest]","Neutral grip Pull-ups"]
pullups =              ["-[Body Weight]-[Bars]---------[5 Reps  - 5 Sets - 30s Rest]","Pull-ups"]
kettlebell_swings =    ["-[10KG]--------[Kettlebell]---[20 Reps - 5 Sets - 1m Rest ]","Kettlebell Swings"]
lateral_raises =       ["-[1.25Kg]------[Dumbbells]----[15 Reps - 5 Sets - 1m Rest ]","Lateral Raises"]
squats =               ["-[10KG]--------[Kettlebell]---[15 Reps - 5 Sets - 1m Rest ]","Squats"]
pushups =              ["-[Body Weight]-[No equipment]-[20 Reps - 5 Sets - 1m Rest ]","Push-ups"]
bar_leg_raises =       ["-[Body Weight]-[Bars]---------[      FAILURE!- 3 Sets     ]","Leg raises"]
med_ball_slam =        ["-[10Kg]--------[Medical Ball]-[      FAILURE!- 3 Sets     ]","med ball slam"]
jumping_jacks =        ["-[Body Weight]-[No equipment]-[      FAILURE!- 3 Sets     ]","jumping jacks"]
floor_workout =        ["-[Body Weight]-[No equipment]-[      FAILURE!- Routine    ]","Floor Workout"]
running_warm_up =      ["-[Body Weight]-[No equipment]-[   simple warmup routine   ]","warmup for running"]
running =              ["-[Body Weight]-[No equipment]-[         Min. 3km          ]","Running!"]
#always train
kettlebell_rolls =     ["-[10KG]--------[Kettlebell]---[      FAILURE!- 3 Sets     ]","Kettlebell Rolls"]
#Warmup train
warm_pullups_pushups = ["-[Body Weight]-[Bars]---------[    FAILURE!- Min.3 Sets   ]","Pull-ups & Push-ups"]

#list of muscles
tricep_muscles = [skull_crushers, dips, dumbbell_kickbacks, diamond_pushups, bench_press]
bicep_muscles = [bicep_curls, hammer_curls, chinups]
forearm_muscles = [kettlebell_rolls, hammer_curls]
back_muscles = [neutral_grip_pullups, chinups, pullups, lateral_raises]
chest_muscles = [dips, diamond_pushups, bench_press, pushups ]
abdominal_muscles = [floor_workout, bar_leg_raises, med_ball_slam, running]
leg_muscles = [kettlebell_swings, squats, jumping_jacks]

#list of prioritized exercises depending on selected muscle group
priority_arms = [bicep_curls, chinups, lateral_raises, dips, pullups]
priority_back = [pullups, lateral_raises, bench_press, neutral_grip_pullups, diamond_pushups]
priority_chest = [pullups, bench_press, dips, diamond_pushups, pushups]
priority_core = [kettlebell_swings, floor_workout, bar_leg_raises, running]
priority_legs = [ squats, kettlebell_swings, floor_workout]

#list of muscle groups
arms = [tricep_muscles, bicep_muscles, forearm_muscles]
back = [back_muscles]
chest = [chest_muscles]
core = [abdominal_muscles] 
legs = [leg_muscles]

specific_muscle_list_names = []
final_exercises_names = []
muscle_group_selection_name= []

print("What muscle group you wanna train today?", "a) Arms", "b) Back", "c) Chest", "d) Core", "e) Legs", sep="\n")
muscle_group_option = input()
if muscle_group_option == "a":
    print(" ", "What part?", "a) Triceps","b) Biceps", sep="\n")
    specific_muscle = input("Select a or b")
    if specific_muscle == "a":
        print(" ")
        print("Main Exercises? Min.1")
        for i in range(len(tricep_muscles)):
            letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]          
            print(f"{letters[i]}) {tricep_muscles[i][1]}")
            specific_muscle_list_names.append(tricep_muscles[i][1])
    if specific_muscle == "b":
        print(" ")
        print("Main Exercises? Min.1")
        for i in range(len(bicep_muscles)):
            letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]          
            print(f"{letters[i]}) {bicep_muscles[i][1]}")
            specific_muscle_list_names.append(bicep_muscles[i][1])
    muscle_group_selection_name.append("arms")

elif muscle_group_option == "b":
    print(" ")
    print("Main Exercises? Min.1")
    for i in range(len(back_muscles)):
        letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]         
        print(f"{letters[i]}) {back_muscles[i][1]}")
        specific_muscle_list_names.append(back_muscles[i][1])
    muscle_group_selection_name.append("back")

elif muscle_group_option == "c":
    print(" ")
    print("Main Exercises? Min.1")
    for i in range(len(chest_muscles)):
        letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]         
        print(f"{letters[i]}) {chest_muscles[i][1]}")
        specific_muscle_list_names.append(chest_muscles[i][1])
    muscle_group_selection_name.append("chest")

elif muscle_group_option == "d":
    print(" ")
    print("Main Exercises? Min.1")
    for i in range(len(abdominal_muscles)):
        letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]          
        print(f"{letters[i]}) {abdominal_muscles[i][1]}")
        specific_muscle_list_names.append(abdominal_muscles[i][1])
    muscle_group_selection_name.append("core")

elif muscle_group_option == "e":
    print(" ")
    print("Main Exercises? Min.1")
    for i in range(len(leg_muscles)):
        letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]          
        print(f"{letters[i]}) {leg_muscles[i][1]}")
        specific_muscle_list_names.append(leg_muscles[i][1])
    muscle_group_selection_name.append("legs")

#loop in order to append main exercises to "final_exercises"
for i in range(len(specific_muscle_list_names)):
    selected_exercises = input("Min. 1 - skip with 'Enter'")
    if selected_exercises == "a":
        final_exercises_names.append(specific_muscle_list_names[0])
    if selected_exercises == "b":
        final_exercises_names.append(specific_muscle_list_names[1])
    if selected_exercises == "c":
        final_exercises_names.append(specific_muscle_list_names[2])
    if selected_exercises == "d":
        final_exercises_names.append(specific_muscle_list_names[3])
    if selected_exercises == "e":
        final_exercises_names.append(specific_muscle_list_names[4])

def fill_exercises_to_target():
    #map muscle groups to their priority lists
    priority_map = {
        "arms": priority_arms,
        "back": priority_back,
        "chest": priority_chest,
        "core": priority_core,
        "legs": priority_legs, 
    }
    
    #get target count
    if muscle_group_selection_name and muscle_group_selection_name[0] == "legs":
        target_count = 3
    else:
        target_count = 5
    
    current_count = len(final_exercises_names)
    
    if current_count < target_count and muscle_group_selection_name:
        muscle_group = muscle_group_selection_name[0]
        
        if muscle_group in priority_map:
            priority_list = priority_map[muscle_group]
            
            #add exercises from priority list
            for exercise_data in priority_list:
                exercise_name = exercise_data[1]
                
                #skip if already in list
                if exercise_name not in final_exercises_names:
                    final_exercises_names.append(exercise_name)
                    current_count += 1
                    
                    #stop when we reach target
                    if current_count >= target_count:
                        break


fill_exercises_to_target()

#print the names
print("\nYour workout plan:")
for i, exercise in enumerate(final_exercises_names, 1):
    print(f"{i}. {exercise}")
print("\nYour workout plan:")

print("=" * 80)

#create a dictionary to easily find exercise details by name
exercise_dict = {
    "Skull Crushers": skull_crushers[0],
    "Dips": dips[0],
    "Dumbbell Kickbacks": dumbbell_kickbacks[0],
    "Diamond Push-ups": diamond_pushups[0],
    "Bench Press": bench_press[0],
    "Bicep Curls": bicep_curls[0],
    "Hammer Curls": hammer_curls[0],
    "Chin-ups": chinups[0],
    "Kettlebell Rolls": kettlebell_rolls[0],
    "Neutral grip Pull-ups": neutral_grip_pullups[0],
    "Pull-ups": pullups[0],
    "Kettelbell Swings": kettlebell_swings[0],
    "Lateral Raises": lateral_raises[0],
    "Squats": squats[0],
    "Floor Workout": floor_workout[0],
    "Hang from bar and leg raises": bar_leg_raises[0],
    "med ball slam": med_ball_slam[0],
    "warmup for running": running_warm_up[0],
    "Running!": running[0],
    "Push-ups": pushups[0],
    "jumping jacks": jumping_jacks[0]
}

#display Warmup
print("Warmup",
      warm_pullups_pushups[1],
      f"   Details: {warm_pullups_pushups[0]}",
      sep="\n")

print("=" * 80)

#display workout with full details
for i, exercise_name in enumerate(final_exercises_names, 1):
    if exercise_name in exercise_dict:
        details = exercise_dict[exercise_name]
        print(f"{i}. {exercise_name}")
        print(f"   Details: {details}")
    print("-" * 80)

#always train part
print(f"6. {kettlebell_rolls[1]}")
print(f"   Details: {details}")

print("=" * 80)
```
